```python
files = glob('內湖西段/直流功率/內湖西段-三期*.csv')
#print(files)

df3 = pd.concat(
    (pd.read_csv(file, usecols=['三期_17-18:17:805:mppt1_power:315'], dtype={'三期_17-18:17:805:mppt1_power:315':float}) for file in files), ignore_index=True)
df3=df3.dropna(axis=0)
print(df3.tail())
```
